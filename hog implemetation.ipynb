```python
import matplotlib.pyplot as plt
import os
import gc
import cv2
import tifffile
import numpy as np
import pandas as pd
from skimage.feature import hog
from skimage import exposure
import cv2
%matplotlib inline
```

```python
def resize_image(img):
    img = cv2.resize(img, (1024, 512))
    gc.collect()
    return img
def load_image(img_path):
    img = tifffile.imread(img_path)
    img = resize_image(img)
    gc.collect()
    return img
```

```python
def load_dataset(df_path, folder_path):
    train_df = pd.read_csv(df_path, usecols = ['image_id', 'label'])
    images, image_labels = [], []
    
    for index in range(len(train_df)):
        img_id, label = train_df.image_id[index], train_df.label[index] 
        img_path = os.path.join(folder_path, img_id + '.tif')
        img = load_image(img_path)
        images.append(img)
        image_labels.append(label)
        print(f"Image {index + 1}/{754} loaded")
        del img, label
        gc.collect()
    return images, image_labels
```

```python
dataset_path = r"E:\compressed_train"
train_df_path = r"E:\train.csv"
```

```python
images, image_labels = load_dataset(train_df_path, dataset_path)
```

```python
def apply_hog(resized_img):
    fd, hog_image = hog(resized_img, orientations=8, pixels_per_cell=(30, 30),cells_per_block=(2, 2), visualize=True, multichannel=True)
    del hog_image
    gc.collect()
    return fd
```

```python
stacked=[]
l=[]
```

```python
for i in range(len(images)):
    stacked.append(apply_hog(images[i]))
    if image_labels[i]=='CE':
        l.append(1)
    else:
        l.append(0)
    print(f"Image {i+1} / 754 stacked")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
std_train=StandardScaler().fit_transform(stacked)
print(std_train)
X_train,X_test,Y_train,Y_test=train_test_split(std_train,l,test_size=0.33,random_state=1)
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
dcc=tree.DecisionTreeClassifier().fit(X_train,Y_train)
```

```python
dcc_predict=dcc.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import ensemble
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
rf=ensemble.RandomForestClassifier(n_estimators=100).fit(X_train,Y_train)
rf_predict=rf.predict(X_test)
knc=KNeighborsClassifier().fit(X_train,Y_train)
knc_predict=knc.predict(X_test)
dcc=tree.DecisionTreeClassifier().fit(X_train,Y_train)
dcc_predict=dcc.predict(X_test)
#print(accuracy_score(Y_test,predict))
print("confusion matrix of RandomForest Classifier: " + str(confusion_matrix(Y_test,rf_predict)))
print("confusion matrix of KNeighbors Classifier: " + str(confusion_matrix(Y_test,knc_predict)))
print("confusion matrix of DecisionTreeClassifier : " + str(confusion_matrix(Y_test,dcc_predict)))
print("Accuracy for RandomForest Classifier: " + str(accuracy_score(Y_test, rf_predict)*100)+"%")
print("Accuracy for KNeighbors Classifier: " + str(accuracy_score(Y_test, knc_predict)*100)+"%")
print("Accuracy for DecisionTreeClassifier : " + str(accuracy_score(Y_test, dcc_predict)*100)+"%")
```
